# Define functions

```python
def get_average(marks):
    np.array(marks)
    return round(np.mean(marks), 2)

def get_exam_average(f):
    # Calculate average of exams
    tot = 50
    gols =  [c for c in f.columns if "exam" in c.lower()]
    exam  = f[gols]
    exam  = exam.apply(lambda x: (x*100)/tot)
    f["avg_exam"]  = exam.apply(get_average, axis=1)
    return  f["avg_exam"]

def get_adaptive_quiz_average(f):
    # Average of "adaptive quiz"
    tot  = 5
    hols = [c for c in f.columns if "adaptive quiz"  in c]
    ad_q = f[hols]
    ad_q = ad_q.apply(lambda x: (x*100)/tot)
    f["avg_ad_quiz"] = ad_q.apply(get_average, axis=1)
    return  f["avg_ad_quiz"]

def get_parp_average(f):
    # Average of participation
    tot = 10
    jols = [c for c in f.columns if "parp" in c]
    parp = f[jols]
    parp = parp.apply(lambda x: (x*100)/10)   
    f["parp"] = parp.apply(get_average, axis=1)
    return f["parp"]

def calculate_total_average(lecture_scores):
    # Return the result based on weightage supplied
    # 65% from exam 
    # 5% from adaptive quiz
    # 20% from quiz
    # 10% from participation
    return round(0.65 * lecture_scores["avg_exam"] +
            0.05 * lecture_scores["avg_ad_quiz"]+
            0.20 * lecture_scores["avg_quiz"]+
                 0.10 * lecture_scores["avg_parp"], 2)
```
